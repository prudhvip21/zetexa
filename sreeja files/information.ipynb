```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd


driver = webdriver.Chrome()  


base_url = "https://www.mwcbarcelona.com/exhibitors"


page_number = 1


wait_time = 20

data = []

try:
    
    driver_info = webdriver.Chrome()  

    while True:
        
        if page_number == 1:
            url = base_url
        else:
            url = f"{base_url}?page={page_number}"
        
        
        driver.get(url)
        
        
        WebDriverWait(driver, wait_time).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr.flex')))
        
        
        soup = BeautifulSoup(driver.page_source, 'lxml')
        fulls = soup.find_all("tr", class_='flex px-4 cursor-pointer hover:bg-gray-100 sm:table-row sm:text-gray-700 sm:font-medium')

        
        for full in fulls:
            company_name_element = full.find("span", class_='ais-Highlight-nonHighlighted')
            company_name = company_name_element.text if company_name_element else "N/A"
            
            featured_element = full.find("span", class_='px-2 py-1 text-xs text-white rounded-full whitespace-nowrap bg-primaryDark')
            featured = featured_element.text if featured_element else "N/A"
            
            stands_element = full.find("div", class_="whitespace-nowrap")
            stands = stands_element.text if stands_element else "N/A"
            
            more_info_element = full["data-href"]
            more_info_link = f'https://www.mwcbarcelona.com{more_info_element}'
            
            
            driver_info.get(more_info_link)
            WebDriverWait(driver_info, wait_time).until(EC.presence_of_element_located((By.CLASS_NAME, 'container')))  # Wait for page to load
            
            
            more_info_content = BeautifulSoup(driver_info.page_source, 'lxml')
            
            
            if more_info_content:
                info_main = more_info_content.find("div", class_="container px-4 py-8 mx-auto md:flex")
                if info_main:
                    info_main1 = info_main.find("main", class_="w-full px-4 md:w-2/3 lg:pr-16")
                    information = info_main.find("div", class_="wysiwyg").text
                    links = info_main.find("ul", class_="w-full")
                    a_tags = links.find_all('a') if links else []
                    links_list = [a['href'] for a in a_tags]
                    
                    
                    print("All links:")
                    for link in links_list:
                        print(link)
                        
                    print("Information:", information)
                    print("links", links_list)
                else:
                    print("No information found for this exhibitor")
            else:
                print("Error: No content found for this exhibitor")
        
        
                
        # Check if the "Next" button is disabled using BeautifulSoup
        next_button = soup.find("li", class_="ais-Pagination-item ais-Pagination-item--nextPage")
        next_page_disabled = next_button is None or "ais-Pagination-item--disabled" in next_button.get("class", [])
        
        # If next page button is disabled, break out of the loop
        if next_page_disabled:
            break
        
        # Increment page number for the next iteration
        page_number += 1

except Exception as e:
    print(f"Error: {e}")

# Close the first WebDriver instance
driver.quit()

# Close the second WebDriver instance
driver_info.quit()
```

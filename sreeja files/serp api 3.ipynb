```python
import requests
from urllib.parse import parse_qsl, urlsplit
import csv


serpapi_key = "c59a7727a5b8b36df3a37d2bfee8aa76e32b3908389d2ad39346628ea168cb3b"
north_west_delhi = [
    "Adarsh Nagar", "Ashok Vihar", "Bawana", "Begum Pur", "Haqiqat Nagar",
    "Karala", "Keshav Puram", "Narela", "Pitam Pura", "Rohini",
    "Rani Bagh, Delhi", "Shalimar Bagh", "Shastri Nagar"
]

north_delhi = [
    "Azadpur", "Civil Lines", "Derawal Nagar", "Gulabi Bagh", "Kamla Nagar",
    "Kashmiri Gate", "Daryaganj", "Model Town", "Narela", "Sadar Bazaar",
    "Sarai Rohilla", "Shakti Nagar", "Tis Hazari", "Timarpur", "Wazirabad",
    "GTB Nagar", "Urdu Bazaar", "Mukherjee Nagar", "Majnu-ka-tilla"
]
search_prefix = "travel agents in "
search_suffix = ", Delhi newdelhi"


search_queries = []
for neighborhood in north_west_delhi:
    search_queries.append(search_prefix + neighborhood + search_suffix)

for neighborhood in north_delhi:
    search_queries.append(search_prefix + neighborhood + search_suffix)
with open("travel_agents_delhi.csv", mode="w", newline="", encoding="utf-8") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["sno", "Title", "rating", "address", "phone", "website"])
    for query in search_queries:
        params = {
            "engine": "google_maps",
            "type": "search",
            "q": query,
            "ll": "@28.63576000,77.22445000,3z",
            "hl": "en",
            "gl": "in",
            "device": "desktop",
            "api_key": serpapi_key,
            "start": "0"
        }

        
        response = requests.get("https://serpapi.com/search", params=params)

        try:
            
            response.raise_for_status()
            
            
            result = response.json()

            
            page_idx = 1
            local_results = result.get("local_results", [])
            for place in local_results:
                title = place.get("title", "N/A")
                rating = place.get("rating", "N/A")
                address = place.get("address", "N/A")
                phone = place.get("phone", "N/A")
                website = place.get("website", "N/A")

                print(f"Result {page_idx}:")
                print(f"Title: {title}")
                print(f"Rating: {rating}")
                print(f"Address: {address}")
                print(f"Phone: {phone}")
                print(f"Website: {website}")
                print()

                
                csv_writer.writerow([page_idx, title, rating, address, phone, website])
                page_idx += 1

            
            while True:
                
                next_page_link = result.get("serpapi_pagination", {}).get("next")
                if next_page_link:
                    next_page_params = dict(parse_qsl(urlsplit(next_page_link).query))
                    next_page_params["api_key"] = serpapi_key

                    response = requests.get("https://serpapi.com/search", params=next_page_params)
                    response.raise_for_status()  
                    result = response.json()
                    local_results = result.get("local_results", [])
                    for place in local_results:
                        title = place.get("title", "N/A")
                        rating = place.get("rating", "N/A")
                        address = place.get("address", "N/A")
                        phone = place.get("phone", "N/A")
                        website = place.get("website", "N/A")

                        print(f"Result {page_idx}:")
                        print(f"Title: {title}")
                        print(f"Rating: {rating}")
                        print(f"Address: {address}")
                        print(f"Phone: {phone}")
                        print(f"Website: {website}")
                        print()

                        
                        csv_writer.writerow([page_idx, title, rating, address, phone, website])
                        page_idx += 1
                else:
                    break

            print("Data has been written to travel_agents_delhi.csv.")
        except requests.exceptions.HTTPError as err:
            print(f"Failed to make the search request for query '{query}': {err}")
            
            csv_writer.writerow([None, f"No results found for '{query}'", None, None, None, None])

print("All queries processed.")
```
